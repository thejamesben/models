## Findings - Comparative Analysis of Model Performance
#### Introduction
Visual comparative analytics provide a clear and concise way to compare the performance of different models. This section presents a script that loads the performance metrics from a CSV file and visualizes these metrics using various plots for comprehensive analysis. The visualizations include bar plots for individual metrics, a heatmap for a comparative overview, and box plots for metric distribution. These visualizations help in understanding the strengths and weaknesses of each model in the context of the specific performance metrics.

#### Loading the Performance Metrics
The script starts by loading the performance metrics from a CSV file into a pandas DataFrame. This step ensures that the data is ready for visualization.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

def load_metrics(file_path):
    # Load the metrics from the CSV file
    metrics_df = pd.read_csv(file_path)
    return metrics_df
```

#### Plotting Performance Metrics
This section of the script defines functions to plot the performance metrics. The functions use seaborn and matplotlib to create bar plots, heatmaps, and box plots for various error metrics (MAE, RMSE, MASE, and WMAPE). The plots are customized with titles, labels, and rotated x-axis labels for better readability.

```python
def plot_performance_metrics(metrics_df):
    # Set the plotting style
    sns.set(style="whitegrid")

    # Plot MAE (Total)
    plt.figure(figsize=(10, 6))
    sns.barplot(x='Model', y='MAE (Total)', data=metrics_df)
    plt.title('Model Comparison - Mean Absolute Error (Total)')
    plt.xlabel('Model')
    plt.ylabel('MAE (Total)')
    plt.xticks(rotation=45)
    plt.grid(True)
    plt.show()

    # Plot RMSE (Total)
    plt.figure(figsize=(10, 6))
    sns.barplot(x='Model', y='RMSE (Total)', data=metrics_df)
    plt.title('Model Comparison - Root Mean Squared Error (Total)')
    plt.xlabel('Model')
    plt.ylabel('RMSE (Total)')
    plt.xticks(rotation=45)
    plt.grid(True)
    plt.show()

    # Plot MASE (Total)
    plt.figure(figsize=(10, 6))
    sns.barplot(x='Model', y='MASE (Total)', data=metrics_df)
    plt.title('Model Comparison - Mean Absolute Scaled Error (Total)')
    plt.xlabel('Model')
    plt.ylabel('MASE (Total)')
    plt.xticks(rotation=45)
    plt.grid(True)
    plt.show()

    # Plot WMAPE (Total)
    plt.figure(figsize=(10, 6))
    sns.barplot(x='Model', y='WMAPE (Total)', data=metrics_df)
    plt.title('Model Comparison - Weighted Mean Absolute Percentage Error (Total)')
    plt.xlabel('Model')
    plt.ylabel('WMAPE (Total)')
    plt.xticks(rotation=45)
    plt.grid(True)
    plt.show()

    # Plot MAE (Seasonal)
    plt.figure(figsize=(10, 6))
    sns.barplot(x='Model', y='MAE (Seasonal)', data=metrics_df)
    plt.title('Model Comparison - Mean Absolute Error (Seasonal)')
    plt.xlabel('Model')
    plt.ylabel('MAE (Seasonal)')
    plt.xticks(rotation=45)
    plt.grid(True)
    plt.show()

    # Plot RMSE (Seasonal)
    plt.figure(figsize=(10, 6))
    sns.barplot(x='Model', y='RMSE (Seasonal)', data=metrics_df)
    plt.title('Model Comparison - Root Mean Squared Error (Seasonal)')
    plt.xlabel('Model')
    plt.ylabel('RMSE (Seasonal)')
    plt.xticks(rotation=45)
    plt.grid(True)
    plt.show()
```

#### Heatmap for Comparative Overview
A heatmap provides a clear and concise comparison of different models across multiple error metrics, allowing for easy identification of the best-performing model for each metric.

```python
def plot_performance_heatmap(metrics_df):
    # Set the plotting style
    sns.set(style="whitegrid")

    # Pivot the dataframe for heatmap
    metrics_pivot = metrics_df.melt(id_vars='Model', var_name='Metric', value_name='Value')
    metrics_pivot = metrics_pivot.pivot('Model', 'Metric', 'Value')

    # Plot heatmap
    plt.figure(figsize=(12, 8))
    sns.heatmap(metrics_pivot, annot=True, fmt=".2f", cmap="YlGnBu", linewidths=.5)
    plt.title('Model Comparison - Performance Metrics')
    plt.xlabel('Metric')
    plt.ylabel('Model')
    plt.show()
```

#### Box Plot for Metric Distribution
Box plots provide insights into the distribution of error metrics across different models, highlighting the variability and central tendency.

```python
def plot_metric_distribution(metrics_df, metric='MAE (Total)'):
    plt.figure(figsize=(10, 6))
    sns.boxplot(x='Model', y=metric, data=metrics_df)
    plt.title(f'Model Comparison - {metric} Distribution')
    plt.xlabel('Model')
    plt.ylabel(metric)
    plt.xticks(rotation=45)
    plt.grid(True)
    plt.show()
```

#### Main Function
The main function orchestrates the loading and plotting of the performance metrics. It calls the load_metrics function to read the metrics from the CSV file and then invokes the various plotting functions to generate the visualizations.

```python
def main(file_path):
    # Load the metrics data
    metrics_df = load_metrics(file_path)

    # Plot the performance metrics
    plot_performance_metrics(metrics_df)
    plot_performance_heatmap(metrics_df)
    plot_metric_distribution(metrics_df, metric='MAE (Total)')
    plot_metric_distribution(metrics_df, metric='RMSE (Total)')
    plot_metric_distribution(metrics_df, metric='MASE (Total)')
    plot_metric_distribution(metrics_df, metric='WMAPE (Total)')
    plot_metric_distribution(metrics_df, metric='MAE (Seasonal)')
    plot_metric_distribution(metrics_df, metric='RMSE (Seasonal)')
```

#### Run Main Function

```python
# Example usage
file_path = 'model_performance_metrics.csv'
main(file_path)
```
