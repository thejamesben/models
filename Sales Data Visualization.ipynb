## Sales Data Visualization
#### Introduction
Visualizing sales data is crucial for understanding trends and patterns over time. This section presents a script that loads sales data from a CSV file and generates two types of plots: monthly sales charts and daily sales charts. These visualizations help in identifying sales trends and making informed business decisions.

#### Loading and Plotting Sales Data
The script starts by loading the sales data from a CSV file into a pandas DataFrame. It then resamples the data to compute monthly and daily sales totals and generates corresponding plots.

```python
import pandas as pd
import matplotlib.pyplot as plt

def plot_sales_charts(file_path):
    # Load the data
    data = pd.read_csv(file_path)
    data['InvoiceDate'] = pd.to_datetime(data['InvoiceDate'])
    
    # Set the InvoiceDate as the index
    data.set_index('InvoiceDate', inplace=True)
    
    # Resample the data to get monthly sales
    monthly_sales = data['TotalPrice'].resample('M').sum()
    
    # Resample the data to get daily sales
    daily_sales = data['TotalPrice'].resample('D').sum()
    
    # Plot monthly sales
    plt.figure(figsize=(12, 6))
    plt.plot(monthly_sales.index, monthly_sales.values, marker='o', linestyle='-', color='b')
    plt.title('Monthly Sales Over Time')
    plt.xlabel('Date')
    plt.ylabel('Total Sales')
    plt.grid(True)
    plt.show()
    
    # Plot daily sales
    plt.figure(figsize=(12, 6))
    plt.plot(daily_sales.index, daily_sales.values, marker='o', linestyle='-', color='r')
    plt.title('Daily Sales Over Time')
    plt.xlabel('Date')
    plt.ylabel('Total Sales')
    plt.grid(True)
    plt.show()
```

#### Running the script
This section  implements the plot_sales_charts function, which loads the sales data from a CSV file and generates the visualizations.

```python
# Usage
file_path = 'preprocessed_data.csv' 
plot_sales_charts(file_path)
```
